```python
# Dependencies
import pandas as pd
import datetime as dt
import warnings
import numpy as np
warnings.filterwarnings('ignore')
```

```python
# Create sample DataFrame1 with columns Matter ID, Name, and Pool Entry Date; and DataFrame2 with Matter ID and Name only
# The idea is we want to create sequence numbers for each Name based on Pool Entry Date (from smallest to largest)
matter_1 = ['A1', 'A2', 'A3', 'B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'C1']
name_1 = ['US333','US333','US333','KR555','KR555','KR555','KR555','KR555','KR555','JP777']
pool_entry = ['1/1/2020','4/1/2020','7/1/2020','4/1/2018','10/1/2018','4/1/2019','7/1/2019','1/1/2020','10/1/2020','10/1/2019']

matter_2 = ['A1', 'B1', 'B2', 'B3']
name_2 = ['US333','KR555','KR555','KR555']

df_1 = pd.DataFrame({'Matter ID': matter_1, 
                     'Name': name_1,
                     'Pool Entry Date': pool_entry})
```

```python
# Create sequence number to assign the version of each matters based on Name and Pool Entry Date (from smallest to largest)
# First, sort data.
df_1["Pool Entry Date"] = pd.to_datetime(df_1["Pool Entry Date"])
df_1 = df_1.sort_values(['Name', 'Pool Entry Date'], ascending=[True, True])

# Second, use cumcount() to number each item in each group from 0 to the length of that group -1. 
# Sequence number will be created in 'Version No.' column
g = df_1.groupby('Name')
df_1['Version No.'] = g.cumcount() + 1
df_1
```

```python
df_1 = df_1.sort_values(['Matter ID'], ascending=[True])
df_1 = df_1.reset_index(drop=True)
df_1
```
